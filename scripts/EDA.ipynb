### Handling the missing values

```python
df.isnull().sum()
```

#### Proportional Allocation
##### If you have the 'circulating_supply' column, you can use it as a proxy for the 'total_supply' column. Calculate the ratio of 'total_supply' to 'circulating_supply' for the available data points and then apply the same ratio to estimate the missing values.

```python
# Calculate the ratio of total_supply to circulating_supply for available data
ratio = df['total_supply'] / df['circulating_supply']
ratio
```

```python
import numpy as np

# Check for infinite value
np.isinf(ratio).any()
```

```python
# above true means we have infinate values in the ratio series
# now we need to replace them with null

# Replace infinite values with NaN
ratio = ratio.replace(np.inf, np.nan)
```

```python
# Check for infinite value
np.isinf(ratio).any()
```

```python
# now we removed the infinate value
# Calutate the average_ratio
average_ratio = ratio.mean(skipna=True)

# fill the nulls with average ratio value with multiply of circulating_supply
# Apply the ratio to estimate total_supply for missing values

df['total_supply'] = df['total_supply'].fillna(df['circulating_supply'] * average_ratio)
```

```python
# after analyzing the above data mostly the max_supply is the round of the total_supply
# so we fill the missing values with the total_supply
df['max_supply'] = df['max_supply'].fillna(df['total_supply'])
```

```python
# as we can see market_cap_change_24h in nan
# we fill that my using below formula
# market_cap_change_24h = (current_price * circulating_supply) - market_cap
# This formula calculates the change in market capitalization over a 24-hour period by subtracting the previous market capitalization (market_cap) from the product of the current price (current_price) and the circulating supply (circulating_supply).

df['market_cap_change_24h'] = df['market_cap_change_24h'].fillna((df['current_price'] * df['circulating_supply']) - df['market_cap'])
```

```python
df.iloc[[458], [9, 10, 11, 12, 13, 14]]
```

```python
# fill the market_cap_change_percentage_24h
# market_cap_change_percentage_24h = (market_cap_change_24h / market_cap) * 100
# This formula calculates the percentage change in market capitalization over a 24-hour period by dividing the market capitalization change (market_cap_change_24h) by the previous market capitalization (market_cap) and multiplying it by 100.

df['market_cap_change_percentage_24h'] = df['market_cap_change_percentage_24h'].fillna((df['market_cap_change_24h'] / df['market_cap']) * 100)
```

```python
df.iloc[[458], [9, 10, 11, 12, 13, 14]]
```

```python
df.isnull().sum()
```

```python
# now i need to fill the fully_diluted_valuation
# fully_diluted_valuation = (current_price * total_supply) / circulating_supply
# This formula calculates the fully diluted valuation by multiplying the current price (current_price) by the total supply (total_supply) and then dividing it by the circulating supply (circulating_supply).

df['fully_diluted_valuation'] = df['fully_diluted_valuation'].fillna((df['current_price'] * df['total_supply']) / df['circulating_supply'])
```
