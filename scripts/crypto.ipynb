```python
import pandas as pd
import requests
```

```python
# API endpoint URL
url = 'https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&order=market_cap_desc&per_page=300&page=1&sparkline=false&locale=en'

# Make a GET request to the API
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Extract JSON data from the response
    json_data = response.json()

    # Create DataFrame from JSON data
    df = pd.DataFrame(json_data)
    
    print("Data successfully fetched..!")
else:
    print("Failed to retrieve data from the API.")
```

```python
# API endpoint URL
url2 = 'https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&order=market_cap_desc&per_page=300&page=2&sparkline=false&locale=en'

# Make a GET request to the API
response2 = requests.get(url2)

# Check if the request was successful
if response2.status_code == 200:
    # Extract JSON data from the response
    json_data_2 = response2.json()

    # Create DataFrame from JSON data
    df_2 = pd.DataFrame(json_data_2)
    
    print("Data successfully fetched..!")
else:
    print("Failed to retrieve data from the API.")
```
